```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
import statsmodels.formula.api as smf


%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format

import warnings
warnings.filterwarnings(action="ignore")

                        
df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
df.head()
```

```python
df["Temperature"]= (df["Temperature (C)"]-df["Apparent Temperature (C)"])
df
```

```python
from sklearn import linear_model

Y = df['Temperature']
X = df[['Humidity','Wind Speed (km/h)' ]]
```

```python
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)
```

```python
print('Değişkenler: \n', lrm.coef_)
print('Sabit değer (bias): \n', lrm.intercept_)
```

```python
import statsmodels.api as sm

X = sm.add_constant(X)
sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

özelliklerin p değerine baktığımda özellikler hedef değişkeni açıklamada istatiksel olarak anlamlı fakat modelimizin R-squared değeri çok düşük yani iyi bir model değil. Kat sayılara baktığımda, humudity Wind speed özelliğine göre hedef değişkeni açıklamada daha güçlü çünkü kat sayısı daha büyük.

```python
df['h_w'] = df["Humidity"] * df['Wind Speed (km/h)']
Y = df['Temperature']
X = df[['Humidity','Wind Speed (km/h)','h_w']]

X = sm.add_constant(X)

sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
# modelimizin R-squared değeri arttı. p değerlerine baktığımda yine anlamlı olduğunu görüyorum fakat yeni özellik eklenmesi
#diğer değişkenlerin katsayılarını düşürdü. Humidity katsayısı diğer modele göre daha da düştü. 
```
